```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
file_path = r"C:\Users\Duong\DSIPractice\visualization\02_activities\assignments\ontario_top_baby_names_female.csv"
data = pd.read_csv(file_path)
data.head() #check data
```

```python
#Plot 1
data_2023 = data[data['Year/Année'] == 2023] #select only 2023 data
top_20 = data_2023.sort_values('Frequency/Fréquence', ascending=False).head(20) #select top 20 from 1993
x_top20 = top_20['Name/Nom']
y_top20 = top_20['Frequency/Fréquence']

fig, ax = plt.subplots(figsize = (8, 4))
bars = ax.bar(x_top20, y_top20, color = 'orange')
ax.set_title('Top 20 female baby names in Ontario 2023')
ax.set_ylabel('Frequency')
ax.set_xlabel('Name')
ax.set_xticklabels(x_top20, rotation=90)
ax.bar_label(bars, fontsize = 6)
```
