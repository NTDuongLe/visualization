```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# create data

x = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])

# define figures and axes

fig, ax = plt.subplots(figsize = (7,3))

# plot bot and plot to ax

ax.bar(x, y1,
       color = 'blue')
ax.plot(x, y2,
        color = 'red')
```

```python
# calculate standard deviation of data

y2_sd = np.std(y2) #standard deviation from numpy

# plot line as before

fig, ax = plt.subplots(figsize = (7,3))

ax.plot(x,y2,
        color = 'red')
```

```python
# add error information using errorbar() - adding some form of error mesaurement is standard practice
fig, ax = plt.subplots(figsize = (7,3))

ax.plot(x,y2,
        color = 'red')


ax.errorbar(x, # our x values
            y2, #our y value
            yerr = y2_sd, # specifies that were plotting vertical error bars
            fmt = 'none') # makes sure we're not plotting the actual data points, only the error
```

```python
# customize error bar appearance

fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x,y2,
        color = 'red')
ax.errorbar(x,y2,
            yerr = y2_sd,
            fmt = 'none',
            ecolor = 'indigo', #error color
            elinewidth = 4, # error linewidth
            capsize = 6, # create caps to the bars
            capthick = 4)
```

```python
# errorevery to change intervals for the error lines
fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x,y2,
        color = 'red')
ax.errorbar(x,y2,
            yerr = y2_sd,
            fmt = 'none',
            ecolor = 'indigo', #error color
            elinewidth = 4, # error linewidth
            capsize = 6, # creates caps to the bars
            capthick = 4,
            errorevery = 2)
```

```python
## Seaborn

# import libraries and packages
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
```

```python
# load tips sample dataset

tips = sns.load_dataset('tips')
print(tips)
```

```python
# create basic plot

sns.lineplot(data= tips, # choose dataset
             x= 'total_bill', #choose x variable
             y= 'tip') #choose y variable
```

```python
# add title and axis labels

tipgraph = sns.lineplot(data=tips,
                        x= 'total_bill',
                        y= 'tip')

tipgraph.set(title= 'Tips vs. Total Bill',
             xlabel= 'Total Bill ($)',
             ylabel= 'Tip Amount ($)')
```

```python
# change figure size

fig = plt.subplots(figsize = (10,3))

tipgraph = sns.lineplot(data=tips,
                        x= 'total_bill',
                        y= 'tip',
                        color= 'blue',
                        linestyle= '--',
                        linewidth= 3,
                        marker= 'o',
                        markerfacecolor = 'yellow')

tipgraph.set(title= 'Tips vs. Total Bill',
             xlabel= 'Total Bill ($)',
             ylabel= 'Tip Amount ($)')
```

```python
# add multiple variables

tipgraph = sns.scatterplot(data= tips,
                           x= 'total_bill',
                           y= 'tip',
                           style = 'time', # set marker style according to time column
                           hue = 'day', # set market hue according to day column, using the palette below
                           palette = ['green', 'indigo', 'deepskyblue', 'yellowgreen'])

tipgraph.set(title= 'Tips vs. Total Bill',
             xlabel= 'Total Bill ($)',
             ylabel= 'Tip Amount ($)')
```

```python
# use pairplot() to compare variables

sns.pairplot(data= tips,
             hue = 'day')
```

```python
# use relplot() to explore specific levels within variables

daysplot = sns.relplot(data= tips, #set data as tips data
                       x= 'total_bill', #set x variable
                       y= 'tip', #set y variable
                       hue= 'sex', #set hue to sex variable
                       col= 'day', # set each axis to different values of day
                       kind= 'scatter', # set plot type to scatter
                       col_wrap= 2) # set number of columns before wrap
```

```python
## Plotly

import plotly.graph_objects as go 

# create data

x1 = np.array(['Luffy', 'Zoro', 'Nami', ' Usopp', 'Sanji'])
y1 = np.array([110,180,240,99,220])
```

```python
## Wordclouds and venn diagrams

# import package
from wordcloud import WordCloud

df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv',
                 on_bad_lines='skip')

df
```

```python
# join all our text from each row from our quote column into a string

text= " ".join(each for each in df.quote) # in each quote join words separated with a space

# generate our wordcloud image

wordcloud = WordCloud(background_color = 'black', colormap= 'inferno').generate(text) # black for accessibility for me

# use matplotlib syntax to put our image in a figure

fig, ax = plt.subplots(figsize = (7,3))
ax.imshow(wordcloud,  # imshow() to show images in axes
          interpolation= 'bilinear') # this line helps smooth our image
ax.axis('off')
```

```python
## Venn Diagrams

#import extension
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
# define our sets

A = set(['apple', 'banana', 'watermelon'])
B = set(['pumpkin', 'blueberry', 'apple', 'key lime'])
```

```python
# create diagram

diagram = venn2([A,B],
                set_labels = ('Fruits', 'Pies'),
                set_colors = ('blue', 'red'),
                alpha = 0.5) # opacity

# show text in diagram

diagram.get_label_by_id('10').set_text('\n'.join(A - B)) # left circle
diagram.get_label_by_id('11').set_text('\n'.join(A & B)) # middle slice
diagram.get_label_by_id('01').set_text('\n'.join(B - A)) # right circle

# id is not ten, eleven, and 1. It's 1 and 0, 1 and 1, and 0 and 1, as in present in left but not right, present in left and right, and present in right but not left, respectively.
# '\n' is so that it creates a new line when joining words
```
